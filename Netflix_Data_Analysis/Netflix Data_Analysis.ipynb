# NETFLIX DATA ANALYSIS


This project performs exploratory data analysis on Netflix movies and TV shows to understand content distribution and trends.


#### PROJECT OBJECTIVE
The objective of this project is to analyze Netflix content data to understand trends in movies and TV shows using exploratory data analysis.

#### IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

#### LOAD THE DATA SET

```python
df = pd.read_csv("D:\\Unified Mentor\\Netflix Data_Analysis\\netflix1.csv")
df.head()
```

#### STEP-1: DATASET OVERVIEW

```python
# SHAPE OF DATASET

df.shape
```

```python
# information of data set

df.info()
```

#### The dataset contains information about Netflix movies and TV shows including title, type, country, release year, rating, duration, and genres.

#### STEP-2: DATA CLEANING

The following cleaning steps were performed:
- Checked missing values
- Filled missing values with 'Unknown'
- Removed duplicate records


```python
# Checking any missing values are present inside the dataset loaded

df.isnull().sum()
```

```python
# Handling the missing values

df.fillna("Unknown", inplace=True)
```

```python
# Remove the duplicates

df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

#### STEP-3: Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is used to understand patterns, trends, and relationships in the dataset.  
In this project, EDA helps analyze Netflix content distribution based on content type, release year, country, rating, and genre.


```python
# Movies vs TV Shows
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

### Movies vs TV Shows

This plot shows the distribution of Movies and TV Shows available on Netflix.

**Observation:**
- Movies are significantly higher in number compared to TV shows.
- This indicates that Netflix focuses more on movie content than series.


```python
#Content Released Over Years

plt.figure(figsize=(8,5))
df['release_year'].value_counts().sort_index().plot()
plt.title("Netflix Content Released Over Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

This above graph represents the number of Netflix titles released each year.

**Observation:**
- There is a steady increase in content after 2010.
- A major growth is visible after 2015, showing Netflixs expansion in content production.


```python
# Top 10 Countries by Content

df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

This bar chart shows the countries with the highest number of Netflix titles.

**Observation:**
- The United States produces the most Netflix content.
- Other major contributors include India, the United Kingdom, and Canada.

```python
# Rating Distribution

sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)
plt.title("Rating Distribution of Netflix Content")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

This visualization shows the distribution of ratings for Netflix movies and TV shows.

**Observation:**
- TV-MA is the most common rating.
- This indicates a large amount of mature content on the platform.

```python
# Top 10 Genres on Netflix

df['listed_in'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

This chart displays the most common genres available on Netflix.

**Observation:**
- Drama is the most popular genre.
- International content and documentaries are also widely available.

## EDA Summary

From the exploratory data analysis, we observe that:
- Netflix has more movies than TV shows.
- Content production increased rapidly after 2015.
- The United States dominates Netflix content production.
- TV-MA is the most frequent rating.
- Drama and international genres are most popular.

This exploratory data analysis provides valuable insights into Netflixs content strategy.
