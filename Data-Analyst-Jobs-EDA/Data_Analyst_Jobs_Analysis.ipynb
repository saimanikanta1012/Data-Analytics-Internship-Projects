This project performs Exploratory Data Analysis (EDA) on a Data Analyst Jobs dataset to identify key hiring trends, industry demand, and job market patterns using data cleaning and visualization techniques.


### IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### LOAD DATASET

```python
df = pd.read_csv("D:\\Unified Mentor\\Data Analyst jobs\\DataAnalyst.csv")
```

### DATA SET OVERVIEW

```python
# Shape of dataset
print("Rows, Columns:", df.shape)
print()
# Column names
df.columns
# Data types and null values
df.info()
```

### Insight:
- The dataset contains 2253 job postings with 16 columns.
- Some columns contain missing or invalid values (-1).


### <CENTER>DATA CLEANING</CENTER>

```python
# Drop unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True)

# Replace -1 with NaN
df.replace(-1, np.nan, inplace=True)

# Check missing values
df.isnull().sum()
```

### <center>Exploratory Data Analysis (EDA)</center>

```python
# COMPANY RATING DISTRIBUTION

plt.figure(figsize=(8,5))
df['Rating'].dropna().hist(bins=20)
plt.xlabel("Company Rating")
plt.ylabel("Number of Job Listings")
plt.title("Distribution of Company Ratings")
plt.show()
```

### Insight:
• Most companies are rated between 3 and 4  
• Very few companies have extremely high ratings  
• Indicates average employer satisfaction


```python
# TOP JOB TITLES

top_titles = df['Job Title'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_titles.plot(kind='bar')
plt.xlabel("Job Title")
plt.ylabel("Number of Jobs")
plt.title("Top 10 Job Titles")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insight:
• "Data Analyst" is the most common role  
• Senior and Reporting Analyst roles show career growth  
• Market demands both entry-level and experienced analysts


```python
# JOB LOCATION ANALYSIS

top_locations = df['Location'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_locations.plot(kind='bar')
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.title("Top Job Locations")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insight:
• Job opportunities are concentrated in tech hubs  
• Urban areas dominate data analyst hiring  
• Location flexibility increases job chances


```python
# INDUSTRY-WISE JOB DISTRIBUTION

top_industries = df['Industry'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_industries.plot(kind='bar')
plt.xlabel("Industry")
plt.ylabel("Number of Jobs")
plt.title("Top Industries Hiring Data Analysts")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insights:
• IT Services and Internet industries lead hiring  
• Analytics is critical in tech-driven sectors  
• Non-tech industries also show growing demand


```python
# COMPANY SIZE VS JOB COUNT

company_size = df['Size'].value_counts()

plt.figure(figsize=(10,5))
company_size.plot(kind='bar')
plt.xlabel("Company Size")
plt.ylabel("Number of Jobs")
plt.title("Company Size vs Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insights:
• Medium and large companies hire more analysts  
• Startups post fewer roles but may offer diverse work  
• Stable organizations invest more in data roles


```python
# COMPANY FOUNDED YEAR ANALYSIS

plt.figure(figsize=(8,5))
df['Founded'].dropna().hist(bins=30)
plt.xlabel("Founded Year")
plt.ylabel("Number of Companies")
plt.title("Distribution of Company Founded Year")
plt.show()
```

### Insights:
• Majority of companies were founded after 1990  
• Newer companies strongly rely on data-driven decisions  
• Data analytics is a modern business necessity


```python
# EASY APPLY VS NORMAL APPLY

easy_apply = df['Easy Apply'].value_counts()

plt.figure(figsize=(6,4))
easy_apply.plot(kind='bar')
plt.xlabel("Easy Apply")
plt.ylabel("Number of Jobs")
plt.title("Easy Apply Job Distribution")
plt.show()
```

### Insights:
• Fewer jobs offer Easy Apply  
• Candidates should prepare for full application processes  
• Resume quality is critical


```python
# SECTION-WISE ANALYSIS

sector_counts = df['Sector'].value_counts().head(10)

plt.figure(figsize=(10,5))
sector_counts.plot(kind='bar')
plt.xlabel("Sector")
plt.ylabel("Number of Jobs")
plt.title("Top Sectors Hiring Data Analysts")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insights:
• Information Technology dominates  
• Finance, Healthcare, and Business sectors follow  
• Cross-domain analytics demand is increasing


### EDA SUMMARY:

1. Data Analyst roles are in high demand across industries
2. Tech and IT sectors dominate job postings
3. Most companies have moderate ratings (3–4)
4. Medium and large companies hire more analysts
5. Urban locations provide more opportunities


###  BUSINESS INSIGHTS
