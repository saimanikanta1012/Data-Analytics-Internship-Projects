<h3 style="font-weight:BOld;color:#001F3F; background-color:#AAAAAA; padding:8px;"> 5.1 OVERALL EMPLOYEE ATTRITION RATE</h3>

```python
# 5 .1 Employee Attrition Count & Rate 
attrition_counts = employee_data['Attrition'].value_counts()

# Create figure
plt.figure(figsize=(16,6))

# Left: bar plot
plt.subplot(1,2,1)
plt.bar(attrition_counts.index,
        attrition_counts.values,
        color=['#ff9999', '#66b3ff'],
        edgecolor='black')
plt.xlabel("Attrition Status", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.title("Employee Attrition Count", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Right: pie chart
plt.subplot(1,2,2)
plt.pie(attrition_counts,
        labels=attrition_counts.index,
        autopct='%1.2f%%',
        startangle=90,
        colors=['#ff9999','#66b3ff'],
        wedgeprops={'edgecolor': 'black'})
plt.title('Employee Attrition Rate', fontsize=14)

# Adjust spacing
plt.tight_layout()
plt.show()
```

### <font color="red">Inference:</font>
1. The Employee Attrition rate of this organization is 16.12%, slightly above the IT industry average (~15%).
2. Indicates a moderate retention risk; requires continuous monitoring.
3. Implement retention strategies like career growth, engagement, and reskilling to control attrition.


<h3 style="font-weight:BOld; color:#001F3F; background-color:#AAAAAA; padding:8px;"> 5.2  Attrition by Gender</h3>

```python
# 5.2 Attrition by Gender

gender_attrition = employee_data.groupby('Gender')['Attrition'].value_counts().unstack()

# Create figure
plt.figure(figsize=(8,5))
gender_attrition.plot(kind='bar',
                      color=['#66b3ff','#ff9999'],
                      edgecolor='black')

plt.xlabel("Gender", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.title("Attrition by Gender", fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Attrition")
plt.show()
```

### <font color='red'>Inference:</font>
* Attrition is slightly **higher among male employees** compared to females.
* However, the difference is **not very significant**.


<h3 style="font-weight:BOld;color:#001F3F; background-color:#AAAAAA; padding:8px;"> 5.3  Attrition by Age Group </h3>

```python
# 5.3 Attrition by Age Group

# Creating bins for age groups
bins = [18,25,35,45,60]
labels = ['18-25','26-35','36-45','46-60']
employee_data['AgeGroup'] = pd.cut(employee_data['Age'], bins=bins, labels=labels, right=False)
age_attrition = employee_data.groupby('AgeGroup')['Attrition'].value_counts().unstack()

ax= age_attrition.plot(kind='bar', stacked=True, figsize=(8,5), color=['#66b3ff','#ff9999'], edgecolor='black')
plt.title('Attrition by Age Group', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
#add value labels at end of bar
for p in ax.patches:
    height=p.get_height()
    x = p.get_x() + p.get_width()/2
    y = p.get_y() + height/2
    if height > 0:
        ax.annotate(
            f'{int(height)}',
            (x, y),
            ha='center', va='center',
            fontsize=9, color='black'
        )
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### <font color="red">Inference:</font>
1. **26-35 age group** shows the **highest attrition** among employees.
2. 36-45 and 46-60 groups have lower attrition rates.
3. Younger employees (18-25) also show a relatively higher attrition rate.

<h3 style="font-weight:BOld; color:#001F3F; background-color:#AAAAAA; padding:8px;"> 5.4 Attrition by Education Field</h3>

```python
# 5.4 Attrition by Education Field

edu_attrition = employee_data.groupby('EducationField')['Attrition'].value_counts().unstack()

plt.figure(figsize=(10,6))
ax= edu_attrition.plot(kind='bar',
                   stacked=True,
                   color=['#66b3ff','#ff9999'],
                   edgecolor='black')

plt.xlabel("Education Field", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.title("Attrition by Education Field", fontsize=14)
plt.xticks(rotation=90)
plt.legend(title="Attrition")
#add value labels at end of bar
for p in ax.patches:
    height=p.get_height()
    x = p.get_x() + p.get_width()/2
    y = p.get_y() + height/2
    if height > 0:
        ax.annotate(
            f'{int(height)}',
            (x, y),
            ha='center', va='center',
            fontsize=9, color='black'
        )

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### <font color="red">Inference:</font>

<h3 style="font-weight:BOld; color:#001F3F; background-color:#AAAAAA; padding:8px;">5.5  Attrition by Job Role</h3>

```python
# 5.5 Attrition by Job Role

# Count attrition for each job role (only employees who left)
job_attrition = employee_data[employee_data['Attrition'] == 'Yes']['JobRole'].value_counts()

# Sort values in descending order for better readability
job_attrition = job_attrition.sort_values(ascending=True)

# Create horizontal bar chart
plt.figure(figsize=(10,5))
bars = plt.barh(job_attrition.index, job_attrition.values, color='#ff9999', edgecolor='black')

# Title and labels
plt.title('Attrition by Job Role', fontsize=14)
plt.xlabel('Number of Employees', fontsize=12)
plt.ylabel('Job Role', fontsize=12)

# Add value labels at the end of bars
for bar in bars:
    plt.text(
        bar.get_width() + 1,                # Position label slightly to the right
        bar.get_y() + bar.get_height()/2,   # Center the label vertically
        str(bar.get_width()),               # Label = value
        va='center', fontsize=9, color='black'
    )

# Grid for x-axis
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### <font color="red">Inference:</font>
* Attrition is **highest among Laboratory Technicians, Sales Executives, and Research Scientists**.  
* These roles may require **better incentives, training, and engagement** to retain employees.

<h3 style="font-weight:BOld;color:#001F3F; background-color:#AAA; padding:8px;"> 5.9 Attrition by Monthly Income</h3>
